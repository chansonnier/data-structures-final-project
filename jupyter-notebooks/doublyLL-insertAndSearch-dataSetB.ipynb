```python
%matplotlib inline
import pylab
import numpy as np
from imageio import imread
```

```python
from matplotlib import pyplot as plt
```

```python
x,y = np.loadtxt('insert_performance_doubly_ll_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

x2,y2 = np.loadtxt('search_performance_doubly_ll_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

plt.plot(y, label = 'Average insertion time')
plt.plot(y2, label = 'Average search time')

plt.legend(loc=0)

plt.title('Avg insert & search time, doubly-linked list, DataSetB')
plt.xlabel('Iteration')
plt.ylabel('Time (nanoseconds)')
plt.ylim(-1000,17500)

F = pylab.gcf()
DPI = F.get_dpi()

plt.savefig('DoublyLLDataSetB.png', dpi=250)
```
