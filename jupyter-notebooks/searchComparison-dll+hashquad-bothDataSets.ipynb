```python
from matplotlib import pyplot as plt
```

```python
x,y = np.loadtxt('search_performance_doubly_ll_dataSetA.csv',
                 unpack = True,
                 delimiter = ',')

x2,y2 = np.loadtxt('search_performance_hashquad_dataSetA.csv',
                 unpack = True,
                 delimiter = ',')


x3,y3 = np.loadtxt('search_performance_doubly_ll_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

x4,y4 = np.loadtxt('search_performance_hashquad_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

plt.plot(y2, label = 'Hash table with open addressing, quadratic probing, Data Set A')
plt.plot(y4, label = 'Hash table with open addressing, quadratic probing, Data Set B')
plt.plot(y, label = 'Doubly-linked list on Data Set A')
plt.plot(y3, label = 'Doubly-linked list on Data Set B')

plt.legend(loc=0)

plt.title('Avg search time for open addressing, quadratic probing \n hash table & doubly-linked list on both DataSets')
plt.xlabel('Iteration')
plt.ylabel('Time (nanoseconds)')

plt.tight_layout()
F = pylab.gcf()
DPI = F.get_dpi()

plt.savefig('searchComparisonBothSets.png', dpi=250, bbox_inches='tight')
```
