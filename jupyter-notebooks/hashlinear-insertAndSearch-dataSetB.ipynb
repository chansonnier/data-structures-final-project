```python
%matplotlib inline
import pylab
import numpy as np
from imageio import imread
```

```python
from matplotlib import pyplot as plt
```

```python
x,y = np.loadtxt('insert_performance_hashlinear_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

x2,y2 = np.loadtxt('search_performance_hashlinear_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

x3,y3 = np.loadtxt('insert_collisions_performance_hashlinear_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

x4,y4 = np.loadtxt('search_collisions_performance_hashlinear_dataSetB.csv',
                 unpack = True,
                 delimiter = ',')

fig, ax = plt.subplots(figsize=(6.5,4))

ax2 = ax.twinx()

curve1, = ax.plot(x,y, label='Average insert times')
curve2, = ax.plot(x2, y2, label= 'Average search times')
curve3, = ax2.plot(x3, y3, label= 'insert collisions per 100 operations', color='r')
curve4, = ax2.plot(x4, y4, label= 'search collisions per 100 operations', color='g')

ax.set_ylabel('Time (nanoseconds)')
ax.set_ylim(-50,800)
ax2.set_ylabel('Number of collisions')
ax2.set_ylim(-5,80)
#ax.set_xlim(0,100)

ax.set_xlabel('Iteration')

plt.legend(loc=1)
ax.legend(loc=2)
plt.title('Avg insert & search time, number of collisions for an\n open addressing, linear probing hash table with DataSetB')
plt.xlabel('Iteration')
plt.tight_layout()
F = pylab.gcf()
DPI = F.get_dpi()

plt.savefig('hashlinearDataSetB.png', dpi=250)
```
